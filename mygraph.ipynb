```python
import matplotlib.pyplot as plt
import numpy as np  
x = []
y = []
z = []
w = []
for line in open('input.txt', 'r'):
    lines = [i for i in line.split()]
    x.append(lines[0])
    y.append(lines[1])
    z.append(lines[2])
    w.append(lines[3])
fig, ax = plt.subplots(4)
ax[0].plot(x, y)
ax[1].plot(y, z)
ax[2].plot(z, x)
ax[3].plot(w, z)
```
